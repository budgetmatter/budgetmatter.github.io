```python
import json
import geojson
from pprint import pprint
import pandas as pd
```

```python
worldgeojson = {
    "type": "FeatureCollection",
    "features": []
}
```

```python
happy = pd.read_csv("map_2019.csv")
happiness = happy.round({'happiness_score': 2, 'gdp_per_capita': 2, 'social_support': 2, 'life_expectancy': 2, 'freedom': 2, 
                'generosity':2, 'government_corr': 2})
happiness.info()
```

```python
happiness.head(25)
```

```python
with open('countries.geo.json') as f:
    features = json.load(f)["features"]
    for foo in features:
        try:
            name = foo['properties']["name"]
            happydata = happiness.loc[happiness["country"] == name, :]
            happydict = happydata.to_dict(orient = 'records')[0]
            foo["properties"] = happydict
            worldgeojson["features"].append(foo)
        except:
            print(f'An error occured for name #: {name}')
            
```

```python
with open('countries.geo.json', 'w') as fp:
    geojson.dump(worldgeojson, fp)
```
